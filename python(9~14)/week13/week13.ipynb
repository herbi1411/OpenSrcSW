```python
#실습1: 생일 기온 시각화
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
high = []
low = []
bdm = "11"
bdd = "15"
dnum = 0
for row in data:
    date = row[0].split('-')
    if row[-1] != '' and row[-2] != '':
        if date[1] == bdm and date[2] == bdd:
            high.append(float(row[-1]))
            low.append(float(row[-2]))
            dnum += 1

plt.plot([i+2011 for i in range(0,dnum)],high,color="pink")
plt.plot([i+2011 for i in range(0,dnum)],low,color="blue")
plt.xticks([i+2011 for i in range(0,dnum)])
plt.show()
```

```python
#실습2: 최고 기온 시각화
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []
jan = []
for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '01':
            jan.append(float(row[-1]))
        elif month == '08':
            aug.append(float(row[-1]))
plt.hist(jan,bins=100,color = 'blue',label = 'Jan')
plt.hist(aug,bins=100,color = 'red',label = 'Aug')
plt.legend(loc="upper right")
plt.show()
```
