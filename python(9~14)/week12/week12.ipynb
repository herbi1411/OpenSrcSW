```python
#실습1: 서울의 일교차
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)
mx = [0,999,-999]
dates = [0] * 3
label = ["일교차가 가장 큰 날","최저기온이 가장 낮은 날","최고기온이 가장 높은 날"]

for row in data:
    row[3] = float(row[3])
    row[4] = float(row[4])
    #일교차
    if row[4] - row[3] > mx[0]:
        mx[0] = row[4] - row[3]
        dates[0] = row[0]
    #최저기온
    if row[3] < mx[1]:
        mx[1] = row[3]
        dates[1] = row[0]
    #최고기온
    if row[4] > mx[2]:
        mx[2] = row[4]
        dates[2] = row[0]

for i in range(3):
    print(label[i])
    print(dates[i])
```

```python
#실습2: 지하철 이용객
import csv
f = open('subwayfee.csv','r')
data = csv.reader(f)
next(data)
subway_line = ["dummy"]
subway_temp = ""
user_list = []
station_list = []
user_list_sorted = []
station_user = 0
station_name = ""
for row in data:
    subway_temp = row[1]
    if subway_temp != subway_line[-1]:
        subway_line.append(subway_temp)
        user_list.append([])
        station_list.append([])
    station_user = int(row[4].replace(",","")) + int(row[5].replace(",",""))
    user_list[-1].append(station_user)
    station_list[-1].append(row[3])
    
del subway_line[0]

for i in range(len(subway_line)):
    print("<",subway_line[i],">")
    user_list_sorted = sorted(user_list[i],reverse = True)
    print("이용객이 가장 많은 역 순위")
    for j in range(3):
        print(j+1,".", station_list[i][user_list[i].index(user_list_sorted[j])],"(",user_list_sorted[j],"명)")
    print("이용객이 가장 적은 역 순위")
    for j in range(3):
        print(j+1,".", station_list[i][user_list[i].index(user_list_sorted[-j-1])],"(",user_list_sorted[-j-1],"명)")
    print()
```
